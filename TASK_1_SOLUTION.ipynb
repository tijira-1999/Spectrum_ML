```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use("ggplot")


# step 1
data = pd.read_csv('student-math.csv', delimiter=';')
df = pd.DataFrame(data)

# step 2
df['final_grade'] = df['G1'] + df['G2'] + df['G3']

# step 3
df = df.drop(["G1", "G2", "G3"], axis=1)

# step 4
df.replace({"yes": "1", "no": "0"}, inplace=True)
# df[["schoolsup","famsup","paid","activities","nursery","higher","internet","romantic"]]

# step 5
fig, ax = plt.subplots()
my_scatter_plot = ax.scatter(
df["studytime"],         # x value
df["final_grade"],       # y value
s = df["studytime"]*40,       # size of each point on the plot
c = df["studytime"],          # colour of each point
cmap = plt.cm.Reds,      # type of colour(colormap = gradient)
vmin = 0,                # lower end of the color gradient
vmax = df["studytime"].max(), # higher end of the color gradient
edgecolor = "#2abb4d",   # edge line color of each dot
linewidth = 0.75         # edge line width of each dot
)
cbar = fig.colorbar(my_scatter_plot)
cbar.set_label("Study time in hours")
ax.set_xlabel("study time")
ax.set_ylabel("final grade")
ax.set_title("Scatter plot")

# scatter plot
plt.show()

# step 6
# box plot
df.boxplot(column=["final_grade", "studytime"])
```
